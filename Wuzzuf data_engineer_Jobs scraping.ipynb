**All Pages** -> there are 87 pages (1304 jobs)

```python
df=pd.DataFrame([],columns=["job_title","company_name","company_page",
                            "location","job_type","skills"])

idx=-1
# there are 87 pages (1304 jobs)
for i in range(0,88):
    
    url=f"https://wuzzuf.net/search/jobs/?a=hpb%7Cspbg&q=Data%20Engineer&start={i}"
    response=requests.get(url)

    soup=BeautifulSoup(response.text,"html.parser")

    Containers=soup.find_all("div",{"class":"css-1gatmva e1v1l3u10"})
    idx=idx+1

    for container in Containers:
        

        job_title=container.find_all("h2",attrs={"class":"css-m604qf"})
        job_title=job_title[0].get_text().strip()
    ####################################################################
        company_name=container.findAll("a",attrs={"class":"css-17s97q8"})
        company_name=company_name[0].get_text().strip()
    ####################################################################
        company_page=container.findAll("a",attrs={"class":"css-17s97q8"})
        company_page=company_page[0].get("href")
    ####################################################################
        location=container.findAll("span",attrs={"class":"css-5wys0k"})
        location=location[0].text.strip()
    ####################################################################
        job_type1=container.findAll("span",attrs={"class":"css-1ve4b75 eoyjyou0"})
        job_type1=job_type1[0].text.strip()

        job_type2=container.findAll("span",attrs={"class":"css-1ve4b75 eoyjyou0"})
        job_type2=job_type2[-1].text.strip()
        try:
            job_type3=container.findAll("span",attrs={"class":"css-o1vzmt eoyjyou0"})
            job_type3=job_type3[0].text.strip()
        except:
            job_type=" / ".join(sorted(list(set([job_type1,job_type2]))))
        else:
            job_type=" / ".join(sorted(list(set([job_type1,job_type2,job_type3]))))

        #job_type=job_type1+" / "+job_type2+" / "+job_type3
    ####################################################################
        skills=container.find_all("div",{"class":"css-y4udm8"})
        skills=skills[0].text.split("·")[1:]
        skills=[s.strip() for s in skills]

        df.loc[idx,"job_title"]=job_title
        df.loc[idx,"company_name"]=company_name
        df.loc[idx,"company_page"]=company_page
        df.loc[idx,"location"]=location
        df.loc[idx,"job_type"]=job_type
        df.loc[idx,"skills"]=skills
                
        idx=idx+1
```
